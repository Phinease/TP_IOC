```python
x = testComplexite["Nb Instance"].tolist()

BestBound = testComplexite["BestBound"].tolist()
DFS10 = testComplexite[" DFS10"].tolist()
DFS01 = testComplexite[" DFS01"].tolist()
BFS = testComplexite[" BFS"].tolist()
Random = testComplexite[" Random"].tolist()

HBestBound = testComplexite["Heur - BestBound"].tolist()
HDFS10 = testComplexite[" Heur - DFS10"].tolist()
HDFS01 = testComplexite[" Heur - FS01"].tolist()
HBFS = testComplexite[" Heur - BFS"].tolist()
HRandom = testComplexite[" Heur - Random"].tolist()
```

```python
x = testComplexiteNBNoeud["Nb Instance"].tolist()

BestBound = testComplexiteNBNoeud["BestBound"].tolist()
DFS10 = testComplexiteNBNoeud[" DFS10"].tolist()
DFS01 = testComplexiteNBNoeud["DFS01"].tolist()
BFS = testComplexiteNBNoeud[" BFS"].tolist()
Random = testComplexiteNBNoeud["Random"].tolist()

HBestBound = testComplexiteNBNoeud["Heur - BestBound"].tolist()
HDFS10 = testComplexiteNBNoeud["Heur - DFS10"].tolist()
HDFS01 = testComplexiteNBNoeud["Heur - FS01"].tolist()
HBFS = testComplexiteNBNoeud["Heur - BFS"].tolist()
HRandom = testComplexiteNBNoeud["Heur - Random"].tolist()
```

```python
def sigmoid(x):
    l = np.log(x)
    return 1/(1 + np.exp(-l))
```

```python
fig, ax = plt.subplots(1, figsize=(15,10))

ax.plot(x, sigmoid(BestBound), label="BestBound Temps")
ax.plot(x, sigmoid(DFS10), label="DFS10 Temps")
ax.plot(x, sigmoid(DFS01), label="DFS01 Temps")
ax.plot(x, sigmoid(BFS), label="BFS Temps")
ax.plot(x, sigmoid(Random), label="Random Temps")

ax.plot(x, sigmoid(HBestBound), label="HBestBound Temps")
ax.plot(x, sigmoid(HDFS10), label="HDFS10 Temps")
ax.plot(x, sigmoid(HDFS01), label="HDFS01 Temps")
ax.plot(x, sigmoid(HBFS), label="HBFS Temps")
ax.plot(x, sigmoid(HRandom), label="HRandom Temps")
```

```python
fig, ax = plt.subplots(1, figsize=(15,10))
ax.plot(x, np.log(x), label = "Log(n)")

ax.plot(x, np.log(BestBound), label="BestBound Nb Noeuds")
ax.plot(x, np.log(DFS10), label="DFS10 Nb Noeuds")
ax.plot(x, np.log(DFS01), label="DFS01 Nb Noeuds")
ax.plot(x, np.log(BFS), label="BFS Nb Noeuds")
ax.plot(x, np.log(Random), label="Random Nb Noeuds")

ax.plot(x, np.log(HBestBound), label="HBestBound Nb Noeuds")
ax.plot(x, np.log(HDFS10), label="HDFS10 Nb Noeuds")
ax.plot(x, np.log(HDFS01), label="HDFS01 Nb Noeuds")
ax.plot(x, np.log(HBFS), label="HBFS Nb Noeuds")
ax.plot(x, np.log(HRandom), label="HRandom Nb Noeuds")

ax.legend()
```
